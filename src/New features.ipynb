```python
def get_emoticon_afflexneg_unigrams(df, df_emoticon_affneg_uni):
    translation_table = dict.fromkeys(map(ord, string.punctuation), None)
    df = df[1].str.lower().str.split()
    afflex_unigram_values = []
    for each_tweet in df:
        word_count = 0
        current_afflex_sum = 0
        considered_words = []
        for each_word in each_tweet:
            considered_words.append(each_word)
            current_series = df_emoticon_affneg_uni[(df_emoticon_affneg_uni[0]==each_word)]
            if current_series.values.size == 0: #Is empty
                continue
            else:
                current_afflex_sum += current_series[1].values[0]
                word_count +=1
        current_afflex_sum = (current_afflex_sum + 1) / (word_count + len(set(considered_words)))
        afflex_unigram_values.append(current_afflex_sum)
    return np.array(afflex_unigram_values).reshape(len(afflex_unigram_values), 1)
emotion_affneg_uni = get_emoticon_afflexneg_unigrams(df, df_emoticon_affneg_uni)
```
